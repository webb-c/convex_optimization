```python
import numpy as np
import matplotlib.pyplot as plt


# Meshgrid
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x,y)
Z = X**2 - Y**2

# Plot the 3D surface
plt.figure(figsize=(5,5),dpi=150)
ax=plt.axes(projection='3d')

ax.plot_surface(X,Y,Z, cmap='seismic')
ax.set_title("$f(x,y)=x^2-y^2$")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
import matplotlib.pyplot as plt
ITER = 30
x0,y0 = -2,-3

# objective function
def f(x,y):
    return x**2 - y**2

# gradient
def grad_x(x,y):
    return 2*x
def grad_y(x,y):
    return -2*y

# learning rate
def alpha(t):
    return 0.1
def beta(t):
    return 0.1

x,y,fxy=[x0],[y0],[f(x0,y0)]

# 1:1 alternating GDA
for t in range(ITER):
    x.append(x[t] - alpha(t)*grad_x(x[t],y[t]))
    y.append(y[t] + beta(t)*grad_y(x[t+1],y[t]))
    fxy.append(f(x[t+1],y[t+1]))

plt.figure(figsize=(5,10), dpi=150)
plt.subplot(2,1,1)
plt.plot(range(ITER+1),fxy,'bo-')
plt.xlabel('$t$')
plt.ylabel('$f(x^{(t)},y^{(t)})$')
plt.subplot(2,1,2)
plt.plot(range(ITER+1),x,'ro-',label='$x^{(t)}$')
plt.plot(range(ITER+1),y,'go-',label='$y^{(t)}$')
plt.xlabel('$t$')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# objective function
def f(x,y):
    return x**2 - y**2

# trajectory
traj = np.array([x,y,fxy])

# Surface grid
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)

# Plot the trajectory
plt.figure(figsize=(5,5),dpi=150)
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, cmap='seismic',alpha=0.5)
ax.plot(*traj,color='black',marker='o',markersize=4,linewidth=1)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$f(x,y)$')
ax.set_title('Trajectory of $(x^{(t)},x^{(t)}, f(x^{(t)},y^{(t)}))$')
plt.show()
```
